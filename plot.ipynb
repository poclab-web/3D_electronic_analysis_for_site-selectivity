```python
import numpy as np
import matplotlib.pyplot as plt

# 入力データ
series_data = {
    'Series A': np.array([[2.18E-02, 2.56E-01],
                          [2.47E-02, 6.25E-01],
                          [5.35E-02, 1.50E+00]]),
    'Series B': np.array([[3.55E-01, 3.78E-01],
                          [8.25E-01, 8.37E-01],
                          [2.04E+00, 2.14E+00]]),
    'Series C': np.array([[2.78E-01, 9.95E-03],
                          [4.78E-01, 1.98E-02],
                          [7.42E-01, 2.96E-02]])
}


colors = ['darkorange', 'royalblue', 'mediumseagreen']

fig, ax = plt.subplots(figsize=(4, 3), facecolor='none')
ax.set_facecolor('none')

# 結果格納用
delta_delta_g_values = {}

# 系列ごとに処理
for i, (label, arr) in enumerate(series_data.items()):
    x = arr[:, 0]
    y = arr[:, 1]

    # 原点通る回帰
    slope = np.sum(x * y) / np.sum(x ** 2)
    y_pred = slope * x

    # 決定係数 R^2（原点回帰）
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r_squared = 1 - ss_res / ss_tot

    # ΔΔG = -1.99 * 0.273 * ln(slope)
    delta_delta_g = -1.99 * 0.273 * np.log(slope)
    delta_delta_g_values[label] = float(round(delta_delta_g,2))

    # プロット描画
    x_fit = np.linspace(0, x.max(), 100)
    y_fit = slope * x_fit
    ax.plot(x_fit, y_fit, color=colors[i], linestyle='-', linewidth=1.5, zorder=0)
    ax.scatter(x, y, facecolors='white', edgecolors=colors[i], s=50, linewidth=1.5, zorder=1, label=label)
    ax.text(x.max(), y_fit[-1]+0.2,
            f'y = {slope:.2f}x\n$R^2$ = {r_squared:.3f}',
            fontsize=8, color=colors[i], verticalalignment='bottom')

ax.scatter(0, 0, color="black", s=10)
ax.set_xlabel(r'$-\log([\mathrm{A}_1]_f/[\mathrm{A}_1]_0)$')
ax.set_ylabel(r'$-\log([\mathrm{A}_2]_f/[\mathrm{A}_2]_0)$')
fig.patch.set_alpha(0.0)
ax.set_xticks(np.arange(4))
ax.set_yticks(np.arange(4))

plt.tight_layout()
plt.show()

delta_delta_g_values
```
